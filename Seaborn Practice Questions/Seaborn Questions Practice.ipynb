```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\HP\Desktop\Cybrom work\Advance Python\Classes Data\Matplot Practice Question\Product.xlsx")
df.head(2)
```

**Exercise 1: Get correlation of columns of
dataset(product.csv) using a seaborn**


```python
pro_corr = df.corr()
pro_corr
```

**Exercise 2: plot the graph using KDE**

```python
# sns.displot(data=pro_corr,kind="kde",fill=True)       # figurelevel Plot

sns.kdeplot(data=pro_corr,fill=True)                    # Axislevel Plot
```

**Exercise 3:plot the graph using Countplot.**

```python
# sns.countplot(data=df,x='toothpaste')                 # Axislevel Plot

sns.catplot(data=df,x='month_number',kind='count')      # figurelevel Plot
```

**Exercise 4:plot the graph using Clusterplot.**

```python
sns.clustermap(data=pro_corr,annot=True)             # Axislevel Plot
```

**Exercise 5: plot the graph using distributed plot
as figure level function.**

```python
sns.displot(data=df,x='facewash',kind="kde",fill=True)
```

**Exercise 6: plot the graph using swarm plot.**

```python
sns.swarmplot(data=df,x='facewash',y='month_number')
```
