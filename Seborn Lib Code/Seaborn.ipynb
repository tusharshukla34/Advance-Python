```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# to get the name of dataset inbuilt in seaborn
print(sns.get_dataset_names())
```

```python
dt = sns.load_dataset('flights')
dt
```

```python
# axis level
sns.scatterplot(data=dt,x='year',y='passengers')
```

```python
sns.lineplot(data=dt,x='year',y='passengers')
```

```python
# figure level
sns.relplot(data=dt,x='year',y= 'passengers', kind = 'scatter')
```

```python
dt = sns.load_dataset('tips')
dt
```

```python
sns.scatterplot(data=dt,x='total_bill',y='tip')
```

```python
sns.scatterplot(data=dt,x='total_bill',y='tip',hue='sex')
```

```python
sns.scatterplot(data=dt,x='total_bill',y='tip',hue='sex',style='smoker')
```

```python
# sns.scatterplot(data=dt,x='total_bill',y='tip',hue='sex',style='smoker',size='size'*20)
sns.lineplot(data=dt,x='total_bill',y='tip',hue='sex',style='smoker',size='size')
```

```python
flt = sns.load_dataset('flights')
flt.head()
```

```python
sns.histplot(data=flt,x='year')
```

```python
sns.kdeplot(data=flt,x='passengers')
```

```python
sns.kdeplot(data=flt,x='passengers',fill=True)
```

```python
sns.kdeplot(data=flt,x='passengers',fill=True,hue='month')
```

```python
# facetplot : figure level
sns.relplot(data=flt,x='month',y='passengers',col='year',col_wrap=4)
```

```python
sns.rugplot(data=flt,x='passengers')
```
