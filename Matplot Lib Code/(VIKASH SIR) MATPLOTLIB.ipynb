### Markers in Matplotlib

In Matplotlib, **markers** are symbols used to highlight individual data points in a plot.  
They are commonly used in functions like `plt.plot()` and `plt.scatter()`.

#### **Basic Example**
```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6], marker='o')
plt.show()

# The `plt.show()`
 function in Python, part of the `matplotlib.pyplot` module, is used to display all figures that have been created but not yet shown.

### Purpose and Usage

#### **Displaying Plots**
After creating a plot using functions like `plt.plot()`, `plt.scatter()`, `plt.hist()`, etc., `plt.show()` is called to render the plot(s) in an interactive window.

### **Context-Dependent Behavior**

#### **1. In a Script**
When running a Python script from the command line, `plt.show()` is essential to open the plot window.  
Without it, the script might execute but no visual output would appear.

#### **2. In an Interactive Environment (IPython, Jupyter Notebook)**
In these environments, plots are often displayed automatically after a plotting command is executed, even without explicitly calling `plt.show()`.  
However, calling `plt.show()` can still be useful to explicitly control when the plot is renderedespecially when creating multiple plots or customizing their display.

# Using xlim and ylim in Matplotlib

xlim and ylim in Matplotlib allow you to manually set the limits of the X-axis and Y-axis in a plot.

### plt.xlim([xmin, xmax])  sets the X-axis range

### plt.ylim([ymin, ymax])  sets the Y-axis range

These are useful for zooming into a specific region or keeping consistent scaling between plots.

Example (use this in a code cell):
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

plt.plot(x, y)

# Set axis limits
plt.xlim(0, 6)   # X-axis from 0 to 6
plt.ylim(0, 10)  # Y-axis from 0 to 10

plt.title("Example of xlim and ylim")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()

Setting both at once

You can also use plt.axis([xmin, xmax, ymin, ymax]):

plt.axis([0, 6, 0, 10])
