```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2D line ploting bivariant analysis hai. 
a = pd.read_csv("cricket.csv")
a
```

```python
plt.plot(a["Year"],a["Rohit"],color="Red",label="Rohit")      # has(#) - hascode ka use hm color ke liye karte hai & esme hm color ko c se bhi lekh sakte hai
plt.plot(a["Year"],a["Virat"],color="Black",label="Virat")
plt.plot(a["Year"],a["Dhoni"],c="#367588",label="Dhoni")  
plt.title("Score Comparision")
plt.xlabel("Year")
plt.ylabel("Score")
plt.legend(loc="upper right")       # Legend() function ka use hm lable ko show karane ke liye karte hai. esme hm attribute ka use bhi kar sakte hai
```

# 10 Dec 2025

```python
# Linestyle - ye 3 type ke hote hai - 1. Dotted  2. Solid  3. Dashed (linestyle bydefault solid hote hai )

plt.plot(a["Year"],a["Rohit"],color="Red",label="Rohit",linestyle="dotted")
plt.plot(a["Year"],a["Virat"],color="Black",label="Virat",linestyle="solid")
plt.plot(a["Year"],a["Dhoni"],c="#367588",label="Dhoni",linestyle="dashed") 
plt.title("Score Comparision")
plt.xlabel("Year")
plt.ylabel("Score")
plt.legend()

plt.show()  # plt.show ka use hm web side bnate hai tb karna padta nhi nhi to hame plot show nhi hote hai. it is use for web application.

```

```python
# Marker - pick point pr jb hame kuch alag show karna hota hai tb hm marker hai use karte hai

plt.plot(a["Year"],a["Rohit"],color="Red",label="Rohit",linestyle="dotted",marker="D")
plt.plot(a["Year"],a["Virat"],color="Black",label="Virat",linestyle="solid",marker="o")
plt.plot(a["Year"],a["Dhoni"],c="#367588",label="Dhoni",linestyle="dashed",marker="*") 
plt.title("Score Comparision")
plt.xlabel("Year")
plt.ylabel("Score")
plt.legend()
```

```python
# Linewidth - eska use hm line ke width ko adjest karne ke liye karte hai.

plt.plot(a["Year"],a["Rohit"],color="Red",label="Rohit",linestyle="dotted",marker="D")
plt.plot(a["Year"],a["Virat"],color="Black",label="Virat",linestyle="solid",marker="o",linewidth=5)
plt.plot(a["Year"],a["Dhoni"],c="#367588",label="Dhoni",linestyle="dashed",marker="*") 
plt.title("Score Comparision")
plt.xlabel("Year")
plt.ylabel("Score")
plt.legend()
```

```python
# Grid - grid ek function hai
plt.plot(a["Year"],a["Rohit"],color="Red",label="Rohit",linestyle="dotted",marker="D")
plt.plot(a["Year"],a["Virat"],color="Black",label="Virat",linestyle="solid",marker="o")
plt.plot(a["Year"],a["Dhoni"],c="#367588",label="Dhoni",linestyle="dashed",marker="*") 
plt.title("Score Comparision")
plt.xlabel("Year")
plt.ylabel("Score")
plt.legend()
plt.grid()
```

```python
product=[2000,3560,9000,8000,7123]
year=[2010,2011,2012,2013,2014]
plt.plot(year,product)
plt.show()
```

```python
# Limitation function - lim() eska use hm x-axis or y-axis dono pr kar sakte hai . esme hame range dena padta hai ke hame kitna 
# data show karna hai graph me without column se data hataye. 
product=[2000,3560,9000,8000,7123,1000000]
year=[2010,2011,2012,2013,2014,2019]
plt.plot(year,product)
plt.ylim(1000,10000)
plt.show()
```

```python
# cmap(color map) :- viridis , PuRd , berlin
# colorbar :- 

year = [2010,2012,2014,2016]
S1 = [78,80,93,88]
S2 = [89,97,95,67]
n = [10,15,18,22]
plt.scatter(year,S1,cmap="viridis",c=n)
plt.colorbar()
```
