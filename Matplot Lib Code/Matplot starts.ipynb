```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2D matploting is a bivarriant analysis
a = pd.read_csv("cricket.csv")
a
```

```python
plt.plot(a['Year'],a['Virat'],color="red",label="Virat")
plt.plot(a['Year'],a['Dhoni'],color="yellow",label="Dhoni")
plt.plot(a['Year'],a['Rohit'],color="green",label="Rohit")
plt.xlabel("Years")
plt.ylabel("Score")
plt.legend(loc= "upper right")  # we use the legend function to show the label in graph
```

```python
                                            # 10-12-2025
```

```python
# linestyle -- It is three types 
# 1. dotted 
# 2. solid
# 3. dashed

plt.plot(a["Year"],a["Virat"],color="red",label = "VIRAT",linestyle="dotted")
plt.plot(a["Year"],a["Rohit"],color="green",label = "ROHIT",linestyle= "solid")
plt.plot(a["Year"],a["Dhoni"],color="yellow",label = "DHONI",linestyle= "dashed")
plt.title("Analysis")
plt.ylabel("Scores")
plt.xlabel("Years")
plt.legend()    # default it will be in upper left


plt.show() # it is used when we create plots for a website or web application. 
              # If we do not use plt.show(), the plot will not be displayed.It is mainly used in web applications.
```

```python
# Marker  We use markers when we want to highlight or show something different at specific data points.

plt.plot(a["Year"],a["Virat"],color="red",label = "VIRAT",linestyle="dotted",marker="D")
plt.plot(a["Year"],a["Rohit"],color="green",label = "ROHIT",linestyle= "solid",marker="o")
plt.plot(a["Year"],a["Dhoni"],color="yellow",label = "DHONI",linestyle= "dashed",marker="*")
plt.title("Analysis")
plt.ylabel("Scores")
plt.xlabel("Years")
plt.legend()   
```

```python
# Linewidth - We can increase the width of the line.

plt.plot(a["Year"],a["Virat"],color="red",label = "VIRAT",linestyle="dotted",marker="D",linewidth=5)
plt.xlabel("Year")
plt.ylabel("Score")
plt.legend()
```

```python
# grid

plt.plot(a["Year"],a["Virat"],color="red",label = "VIRAT",linestyle="dotted")
plt.plot(a["Year"],a["Rohit"],color="green",label = "ROHIT",linestyle= "solid")
plt.plot(a["Year"],a["Dhoni"],color="yellow",label = "DHONI",linestyle= "dashed")
plt.title("Analysis")
plt.ylabel("Scores")
plt.xlabel("Years")
plt.grid()
plt.legend()    
```

```python
# multigraph

year = ["2010","2015","2020"]
score = [[98,37,90],[10,76,54],[58,99,90]]

# plt.bar(year,score[0],fc="red",ec="black",width=0.4)
# plt.bar(year,score[1],fc="violet",ec="black",width=0.4)
# plt.bar(year,score[2],fc="purple",ec="black",width=0.4)
x =np.arange(len(year))
w = 0.25
plt.bar(x - w,score[0],fc="red",ec="black",width=w)
plt.bar(x,score[1],fc="violet",ec="black",width=w)
plt.bar(x + w,score[2],fc="purple",ec="black",width=w)
plt.xticks(x,year)
plt.show()
```

```python
# Stacked Graph

countries = ["USA","India","China","Russia"]
gold = [56,76,56,100]
silver = [120,99,128,80]
bronze = [120,198,111,200]

w=0.5
b = np.add(gold,silver)

plt.bar(countries,gold,w,color="gold",label="GOLD",ec="red")
plt.bar(countries,silver,w,color="silver",bottom=gold,label="SILVER",ec="black")
plt.bar(countries,bronze,w,color="brown",bottom=b,label="BRONZE",ec="black")
plt.xlabel("<------------------ Countries ------------------>")
plt.legend()
plt.show()
```
