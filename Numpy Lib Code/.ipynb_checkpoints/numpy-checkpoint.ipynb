```python
import numpy as np
import time as t

size = 10
list1 = range(size)
list2 = range(size)
start = t.time()
result = [x+y for x,y in zip(list1,list2)] # zip() is used to pair items from two or more lists together, position by position.
                                           # list1 = [1, 2, 3]
                                           # list2 = [4, 5, 6]
                                           # → zip gives: (1,4), (2,5), (3,6)
                                           # List Comprehension
end = t.time()
print("total time taken by list", end - start)

arr1 = np.arange(size)
arr2 = np.arange(size)
start = t.time()
result = arr1 + arr2
end = t.time()
print("total time taken by array", end - start)
```

```python
# problem
x = np.array([[1,3,5,6,7,8]])

for i in x:
    for j in i:
        print(j)
```

```python
# solution
# nditer make ease to access element of array dimention do not need to write many loops
for i in np.nditer(x):
    print(i)
```

```python
# tells indexing with nditer
#ndenumerate
x = np.array([[1,3,5,6,7,8]])
for i,d in np.ndenumerate(x):
    print(i,d)
```

```python
x = np.array([[[1,3,5,6,7,8]]])

for i,d in np.ndenumerate(x):
    print(i,d)
```

```python
# flatten used o chane the dimention of array 2d to 1d

n =np.array([[2,3],[4,6],[5,6]])
d = n.flatten()
print(d)
d[1]=600
print(d)
print(n)     # see the original values do not get change even i assign a new value above
```

```python
# # ravel used o chane the dimention of array 2d to 1d

k = np.array([[2,3],[4,6],[7,8]])
k
h = k.ravel()
print(h)
h[1] = 390
print(h)
print(k)    # see the original values get changed when i assign a new value above
```
