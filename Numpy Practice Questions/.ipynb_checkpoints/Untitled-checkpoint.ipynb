```python
a = np.random.randint(1,10,25).reshape(5,5)
a
```

```python
# flatten

ab = a.flatten()

ab[7] = 45
ab

# ravel 
ab = a.ravel()

ab[7] = 888
ab
a[4,3] = 9000
a
ab
```

```python
l = [2,3,5,6,8]
lt1 = np.array(l)

l2 = [24,43,45,46,48,78]
lt2 = np.array(l2)
```

```python
h = np.concatenate([lt1,lt2])
```

```python
np.split(h,5)
```

```python
np.array_split(h,3)
```

```python
np.sort(h),ascending=False)
```
